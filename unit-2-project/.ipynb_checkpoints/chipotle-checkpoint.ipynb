## Intermediate Level

### Part 3: Calculate the average price of an order.

_Hint:_ Examine the data to see if the `quantity` column is relevant to this calculation.

_Hint 2:_ String methods can be called via `str`. For example, you might find the following useful:

```python
data.variable.str.replace('xyz', 'abc')
```

_Hint 3:_ Pay close attention to data types here! This is tricky.

````python
# First get create a column without the '$' in front of the price strings and cast those prices as
# floats
orders['item_price_float'] = orders['item_price'].str.replace('$', '').astype(float)
# Per advice from Tim and Baur - check for item prices that have had their  values increased 
# by the quantity # already
# First just listing medians (mean would most likely be affected by the bad entries)
orders.groupby('item_name')['item_price_float'].agg(['median','mean'])
````

````python
# Get all columns whose 'item_price' is closer in size to the 
# (median price x the number of items ordered) than the
# (median price of one item)
# save this boolean series to determine how to calculate total price in the next block
# 'groupby().get_group()' from:
# https://stackoverflow.com/questions/14734533/how-to-access-pandas-groupby-dataframe-by-key
already_mult = orders.apply( lambda x: (x['quantity'] > 1) & 
                    (abs(x['item_price_float'] - x['quantity'] * 
                    orders.groupby('item_name').get_group(x['item_name'])['item_price_float'].\
                    median()) < abs(x['item_price_float'] - 
                    orders.groupby('item_name').get_group(x['item_name'])['item_price_float'].\
                    median())), axis=1)
````

````python
# Next create a column with the total cost of every item listed - initialize to float all zeros
orders['item_price_total'] = 0.0
# If the price already acounts for the number of items ordered, make the total price just the 
# existing float
orders.loc[already_mult, 'item_price_total'] = orders['item_price_float']
# Otherwise take the item_price and multiply it by the quantity
orders.loc[~already_mult, 'item_price_total'] = orders['item_price_float'] * orders['quantity']
# Verify that you have the correct data
orders.head()
````

````python
# Group the data by order, sum all of the items in each order and return the average of those 
# totals
cost_of_orders = orders.groupby('order_id')['item_price_total'].sum()
cost_of_orders.mean()
````

## Advanced Level


### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


````python
# Make use of the cleaned up row to create dataframe of burrito orders and
# topping choices
# '.str.contains' from:
# https://pandas.pydata.org/pandas-docs/version/0.17.0/generated/pandas.Series.str.contains.html
burrito_toppings = orders.loc[orders['item_name'].str.contains('Burrito'),
                              ['choice_clean']]
````

````python
# Create a column with the toppings as a list by splitting the strings
# by ','
burrito_toppings['choice_list'] = burrito_toppings['choice_clean'].str.split(',')
````

````python
# Use apply to get the lengths of all of the toppings lists and put that
# into a new column
burrito_toppings['choice_length'] = burrito_toppings['choice_list'].apply(
    lambda choice_list: len(choice_list))

# Get the mean of the topping list-lengths column
burrito_toppings['choice_length'].mean()
````

### Part 6: Create a dictionary. Let the keys represent chip orders and the values represent the total number of orders.

Expected output: `{'Chips and Roasted Chili-Corn Salsa': 18, ... }`

Note: Please take the `quantity` column into account!

Optional: Learn how to use `.defaultdict()` to simplify your code.

````python
# Get the orders that have 'Chip' in the item name - assign to a new dataframe
# that has the item names and their order quantities
chip_orders = orders.loc[orders['item_name'].str.contains('Chip'),
                              ['item_name', 'quantity']]
# '.to_dict' from looking up '.defaultdict()':
# https://stackoverflow.com/questions/42771719/how-to-convert-pandas-dataframe-to-defaultdict-class-list-where-one-of-the-co
# using '[quantity]' at the end because I found that the '.to_dict()'
# function put the whole dataframe dictionary as a value to the key
# 'quantity'
chip_dict = chip_orders.groupby('item_name').sum().to_dict()['quantity']
# Print out the dictionary to verify the output
chip_dict
````

### Part 7: Create a barplot of the frequencies of the chip orders from part 6.

````python
# Use the 'chip_orders' dataframe variable from the previous problem
# group the dataframe by separate item names, get the order-frequency sum, and sort them
# by order quantity (for easier plot reading)
group_data = chip_orders.groupby('item_name').sum().sort_values(by = 'quantity')
#  reset the grouped data index so that I can use the 'item_name' variable as different
# groups that allow for auto-coloring of bars
# '.reset_index' information from:
# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.reset_index.html
group_data.reset_index(inplace=True)
# Use the '.plot()' command, with 'kind' as 'bar'
# Use a larger plotsize for easier plot reading
# Default behavior of this plot was to have a legend for the bar color
# This is unnecessary since it is one 'series' and doesn't
# supply colors for all different orders
# and all of the bars are labeled with their orders
# Source: https://stackoverflow.com/questions/20865487/pandas-python-plot-without-a-legend
group_data.plot(kind = 'bar', x='item_name', y='quantity', legend = False, figsize = (12, 12));
plt.xlabel('Chip Orders')
plt.ylabel('Frequency')
plt.title('Frequency of Chip Orders');
````

````python
# 1. What are the 10 most revenue-generating item?
# group dataframe by item name and get per-item revenue using the item price
# column data from Part 3 of the notebook
# Sort these values from highest to lowest
item_revenue = orders.groupby('item_name')['item_price_total'].sum().sort_values(ascending = False)
# Show the 10 items with the highest revenue
item_revenue.head(10)
````

````python
# 2. What is the largest "profit-margin" item? In this case, looking for most-ordered item 
#    with the highest price

# Create an item_count of all items sold
item_count = orders['item_name'].value_counts()
# Create a Dataframe using the revenue-total dataframe from
# the previous code block and the item-count dataframe
# just created by concatenating the two together as columns 
# ('axis = 1'). 'sort = True' from Pandas warning to ensure proper 
# alignment of indeces (item names in this case)
finance_df = pd.concat([item_revenue, item_count], axis = 1, sort = True)
# rename the columns for easier reading
finance_df.columns = ['item_revenue', 'item_count']
# Create an 'average cost' column because the items can cost different amounts based on
# the sides ordered with them
finance_df['avg_item_price'] = finance_df['item_revenue'] / finance_df['item_count']
# Verify that the dataframe looks the way we want
finance_df.head()
````

````python
# Function to go through every row and compare cost and number of times sold, saving the item_name (index), price,
# and frequency of the maximum combo
def best_combo(df = finance_df, col_1 = 'item_count', col_2 = 'avg_item_price'):
    max_combo = ('', 0, 0)
    for index, row in df.iterrows():
        if row[col_1] > max_combo[1] and row[col_2] > max_combo[2]:
            max_combo = (index, row[col_1], row[col_2])
    return max_combo

best_combo()
            
````

````python
# 3. Are there any repeat orders (completely)? Which may be indicative of repeat-customers
#    or at least a very popular combo
# '.value_counts()':
# http://pandas.pydata.org/pandas-docs/version/0.17/generated/pandas.Series.value_counts.html
# add commas on to end of item_names in new columns to make summation easier to read, using '.apply()' for this
orders['item_name_spaces'] = orders['item_name'].apply(lambda item: item + ', ')
# Remove end ', ' from each summed order for cleaner look and count number of total order 
# occurences
# Once all of the items per order are summed together as single strings, count the number of occurences
# for each total order and return the Series, indicating the number of times each
# combination was ordered
orders.groupby('order_id')['item_name_spaces'].sum().apply(lambda item: item[:len(item) -2]).value_counts()
````
