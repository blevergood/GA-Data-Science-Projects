### Challenge 1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# Put logic into a function to make the solution generalizable.
# Inputs to the function will be the upper- and lower-bound of the range that you want to use as 
# factors to find the largest palindrome
# (100, 999) will be used as the inputs to solve the specific problem statement.
def get_palindrome(min_number, max_number):
    # Initialize variables to store important values
    # The two factors & the palindrome product that we want to return
    factor_1 = 0
    factor_2 = 0
    palindrome = 0
    # Go through every number in our defined range, and check it against every number in range
    # This verifies that you've seen all possible products - we're using the range() function
    # to iterate through all desired values (100 through 999)
    for factor_a in range(min_number, max_number + 1):
        for factor_b in range(min_number, max_number + 1):
            # Perform a calculate once per inner-loop so that both factors are defined
            hold = factor_a * factor_b
            # To decide if we want to keep a product:
            #     1. In order to see if the product is a palindrome, compare it to it's reverse
            #     This can only be done with strings
            #     String-reversing function found at: 
            #         https://stackoverflow.com/questions/931092/reverse-a-string-in-python
            #     2. Additionally, verify that the product is larger than your current palindrome
            #     value. Otherwise we don't want to return it
            if str(hold) == str(hold)[::-1] and hold > palindrome:
                # If we have found a larger palindrome, overwrite our return values with 
                # the current factors and product
                factor_1 = factor_a
                factor_2 = factor_b
                palindrome = hold
    # Return the eventual champions as a tuple of values
    return factor_1, factor_2, palindrome 

# In order to solve the problem as stated, assign return values from function above using 100-999
# as the range via the two input values and print the return values out
factor_1, factor_2, palindrome = get_palindrome(100, 999)
print("The Palindrome is: " + str(palindrome) + " = " + str(factor_1) + " x " + str(factor_2))
```


### Challenge 2: Summation of Primes
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below 2,000. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# From Wikipedia: "Given an input number n, check whether any prime integer m from 2 to √n evenly
# divides n."
# If one does, the number 'n' is not prime, if it doesn't, 'n' is prime
# Source: https://en.wikipedia.org/wiki/Primality_test
import math # need to import this module to use the sqrt() function
# Create a function with an input variable to make the function generalizable.
# The input function will be the largest number that you want to evaluate for prime-ness before
# summing results.
# The (1999) will be used as the input to solve for the problem statement
def sum_primes(max_number):
    # Keep a running list of prime numbers found, 2 is the first prime number by definitiion, 
    # so we can just start with it in our list
    primes = [2]
    
    # Check every number above 2 in the input range for prime-ness using the strategy described
    # above
    for number in range(3, max_number + 1):
        #assume each number is prime...makes the logic is easier to write for finding the 
        # NOT cases. This will be the evaluation variable for whether or not we add it to our 
        # list.
        is_prime = True
        # For every number being evaluated, perform the check defined at the top:
        #     Go through every prime number already collected that is less than 
        #     sqrt(the_number_being_evaluated) and check whether the number being 
        #     evaluated can be devided by any of those prime numbers evenly
        #     (i.e. with a remainder of zero - the modulo operator is perfect for this).
        #     If yes, the number isn't prime, don't add it to our list
        #     If no, it's prime - add it to our list
        for prime in primes:
            # Stop evaluating the current number if the prime is above its's square root, 
            # no need to keep going
            if prime > math.sqrt(number):
                break
            # If we determine that a number isn't prime, stop evaluating, no need to keep going
            if number % prime == 0:
                is_prime = False
                break # And no need to keep evaluating
         
        # If we didn't find a 'False' case, it must be true
        # Add this number for the next round
        if is_prime:
            primes.append(number) 
    # Return the list of primes and the sum of that list after the loop
    return primes, sum(primes)

# assign the proper output to variables using 1999 as the input above, and
# print those outputs to fulfill the specific problem statement
prime_list, prime_sum = sum_primes(1999)
print("The sum of all primes found under the number 2000 is: " + str(prime_sum))
print("The list of primes found under the number 2000 is:\n" + str(prime_list)) 
```

### Challenge 3: Multiples of 3 and 5
If we list all of the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6, and 9. The sum of these multiples is 23. Find the sum of all the multiples of 3 and 5 below 1,000. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# Create a function that takes in 3 variables to make this function generalizeable
# The three functions will be the first number that we want to check for divisibility against,
# the second number that we want to check for divisibility against, and the maximum number that we 
# want to check for divisibility.
# The inputs will be 999, 3, and 5 in order to solve for the problem statement
def sum_multiples(limit, multiple_a, multiple_b):
    # Keep track of the numbers that are divisible by your factoring inputs using a list
    nums_add = []
    # For every number in your defined range, check if they are divisible by your factoring inputs
    # If they are, add them to your tracking list. Using the OR operator so we get all numbers
    # that are divisible by either one of, or both of the factoring inputs
    for number in range(1, limit+1):
        if number % multiple_a == 0 or number % multiple_b == 0:
            nums_add.append(number)
    # return the list of divisible numbers and the sum of the list after the loop
    return nums_add, sum(nums_add)

divisible_nums, num_sum = sum_multiples(999, 3, 5)
print('The sum of all the numbers under 1000 that are multiples of 3 or 5 is: ' + str(num_sum))
print('The list of the numbers found under 1000 that are multiples of 3 or 5 is:\n' + str(divisible_nums))
```

### Challenge 4: String Compressor
Implement a method to perform basic string compression using the counts of repeated characters. (This is called run-length encoding.) For example, the string "aabcccccaaa" would become a2b1c5a3. If the “compressed” string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a–z). Specify whether your solution is case sensitive or case insensitive and what you would need to change to make it the other. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# Original method - pre-recursion challenge
# Create a function with a single input being the string to be compressed.
def string_compress(string):
    # Using a dictionary-like concept with a list of letters for every new, letter sequence found
    # and a list for the count of each of those letters (can't use a dictionary
    # because you can't have multiples of keys)
    letters = [] 
    counts = []
    # create a string to hold compressed string as it's being built
    zip_string = ''
    # Need to be able to index the string, can use the range function with the length of the input
    # string as the upper bound, the loop will walk from 0 through len(string) - 1
    for index in range(len(string)): 
        # The first letter doesn't need to be evaluated, just add it's letter and 1 to our lists
        if index == 0: 
            letters.append(string[index])
            counts.append(1)
        else:
            # If our letter is the same as the previous one, increment the count
            if string[index] == string[index - 1]:
                counts[-1] += 1
            # Otherwise it's a new letter squence.
            # Add the last letter in the letters list (the previous letter), and the last count
            # in the counts list (the corresponding count) to our eventual output string and 
            # start over by adding the new letter and a starting count of '1' to the respective 
            # lists
            else: 
                zip_string = zip_string + letters[-1] + str(counts[-1])
                letters.append(string[index])
                counts.append(1)
    # After the above for-loop completes, you will need to offload the final letter/count combo
    # onto the output string
    zip_string = zip_string + letters[-1] + str(counts[-1])
    
    # Once we've created the compressed output string, verify that it is longer than the original 
    # string. If it is, return it, otherwise allow the function to return the original string
    if len(string) <= len(zip_string):
        return string
    return zip_string

# Use the given example from the problem statement, and a short string to show that the compressor 
# works appropriately in both cases
# Got concept for breaking up print statement into multiple lines from: 
# https://stackoverflow.com/questions/4172448/is-it-possible-to-break-a-long-line-to-multiple-lines-in-python
print( "Using the example string from the problem statement: 'aabcccccaaa', ",
      "the compressed string output is: '" + string_compress("aabcccccaaa") + "'")
print("Using a short string: 'abcd', the compressed string output is: '" + string_compress("abcd") + "'")

# This solution is Case-Sensitive, I would have to use a .lower() or .upper() on the input string
# to make the function Case-Insensitive. I could perform these operations on every single indexing
# of each input string, but it would be faster to do it once at the begining
```

```python
# For the recursion version of the problem, I'm starting with the code that we developed in class
# for inspiration. i.e. remove characters from a string copy of the original so we can crawl 
# through the string and grab our letters and letter-counts as we go
def class_compress_code(word):
    # Create a copy of the input word so we can save the original for our final length comparison
    word_copy = word
    # initialize our output string to start compiling our compressed input
    str_out = ""
    # Keep evaluating and building the compressed string as long as our word copy has letters
    # remove the letters from the previous same-letter sequence each time we find a new letter 
    # and need to start over with counting
    while len(word_copy) > 0:
        # Create a holding variable for the letter that we are checking against and initialize a 
        # counter for that letter
        first = word_copy[0]
        counter = 0
        # iterate through the letters in the word copy (the one being truncated at the end of
        # every loop) and add one to the counter if it is
        for char in word_copy:
            if char == first:
                counter += 1
            # if we find a mismatch, it's time to stop evaluating because we don't want to lump
            # together separate runs of the same letter
            else:
                break
        # After the loop add the current letter being evaluted and it's count to the eventual 
        # output string and truncate the word copy for the next loop so you get rid of the
        # letters that you had just counted
        str_out = str_out + first + str(counter)
        word_copy = word_copy[counter:]
    # TODO: Ensure output is <= original length (Done)
    # Same logic as the above statement
    if len(word) <= len(str_out):
        return word
    return str_out
# Print the outputs using the same inputs from before to verify that it works by comparing outputs
print(class_compress_code("aabcccccaaa"))
print(class_compress_code("abcd"))
# Again this solution would be Case-Sensitive.  I would have to use .lower() or .upper() on the
# input string at the beginning to make it Case-Insensitive
# I wrote out this portion of the code with comments to show that I understand how it works and I 
# can use it in the following recursive solution to this problem
```

```python
# Creating a recursive function with a 'base case' of input word has a length of zero, that will
# cause the function to begin completing and 'rising up' through the previous recursions once 
# we've created the full compressed string (and have walked through the entire input string)
# using a 'level' variable set to a default of '0' so we know when we're back at the top of the 
# function to perform our final length evaluation before giving the final output from the function
def recursive_compress(word, level = 0):
    str_out = ''
    # If the length of the input word is 0, which python equates to False, return the output
    # string, because we've compiled the last portion of it.
    # The return statement will return us to the previous level of the function call until we exit
    # out of the first function call with our final value
    if not word:
        return str_out
    # If we have an input string to evaluate, use the same logic from class:
    # grab the first letter of the string and intialize a counter for that letter
    first = word[0]
    counter = 0
    # iterate through the letters in the word copy (the one being truncated at the end of every 
    # loop) and add one to the counter if it is
    for char in word:
        if char == first:
            counter += 1
        # if we find a mismatch, it's time to stop evaluating because we don't want to lump 
        # together separate runs of the same letter
        else:
            break
    # After the loop, add the letter being evaluated and it's final count to our output string
    str_out = str_out + first + str(counter)
    # Evaluate what the output string would be at this level by calling the function with the 
    # shorter version of the string. increment the 'level' variable with each recursive call so 
    # we know when not to evaluate original and compressed string lengths.
    # Don't return it so that we have the ability to compare the length of the final product with
    # our original input
    str_out = str_out + recursive_compress(word[counter:], level + 1)
    # After we've gotten our final output from a given level of the function, if that level is the
    # first call of the function (level = 0) which Python evaluates to False, then perform a 
    # length-check against the original string and final compressed string, which will both be 
    # present in this top level of the function.
    if not level:
        # If we're at the top level and the compressed string isn't smaller than the original
        # return the original word
        if len(str_out) >= len(word):
            return word
    # Otherwise f we're not at the top level, or our original word is larger than the compressed 
    # string, return it
    return str_out

# Print the outputs using the same inputs from before to verify that it works by comparing outputs
print( "Using the example string from the problem statement: 'aabcccccaaa', ",
      "the compressed string output is: '" + string_compress("aabcccccaaa") + "' using recursion")
print("Using a short string: 'abcd', the compressed string output is: '" + string_compress("abcd") + "' using recursion")
# Finally, this solution is also case-sensitive since it uses logic from the solution above.
# I would need to utilize the .upper() or .lower() operators on the input string to make the
# problem Case-Insensitive
```

### *BONUS* Challenge: FizzBuzz
Write a program that prints all of the numbers from 1 to 100. For multiples of 3, instead of the number, print "Fizz;" for multiples of 5, print "Buzz." For numbers that are multiples of both 3 and 5, print "FizzBuzz." Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# This problem is very similar conceptually to Problem 3 - will use the modulo divider operation
# again, but not in the same logic line for easier String manipulation

# Create a function with 4 inputs, the smallest and largest numbers in the range you are checking
# and printing and the two numbers against which you would like to check to cause the function to 
# print out 'Fizz', 'Buzz', or 'FizzBuzz'
# The inputs to the function will be: 1, 100, 3, and 5 to fit the specific problem statement
def fizz_buzz_counter(min_number, max_number, fizz, buzz):
    # Have a running string to print out the word variants
    fizz_buzz = ''
    # Check every number at the beginning to see if it is divisible by the 'Fizz' number 
    # or the 'Buzz' number.
    for number in range(min_number, max_number + 1):
        fizz_buzz = ''
        # Use separate if statements and put the 'Fizz' evaluation first so that you properly add 
        # 'Fizz' to your running string (fizz_buzz) before 'Buzz' in the case that the number 
        # being checked is divisible by both
        if number % fizz == 0:
            fizz_buzz += 'Fizz'
        if number % buzz == 0:
            fizz_buzz += 'Buzz'
        # After both evaluations, check if the running string as length greater than 0 (is 'True')
        # if it is, print the string (which will be 'Fizz', 'Buzz', or 'FizzBuzz'), 
        # otherwise just print the number that is being evaluated
        if fizz_buzz:
            print(fizz_buzz)
        else:
            print(number)
# Got concept for breaking up print statement into multiple lines from: 
# https://stackoverflow.com/questions/4172448/is-it-possible-to-break-a-long-line-to-multiple-lines-in-python
print('The following are the results of the FizzBuzz counter operation with a range from', 
      ' 1 through 100, \nchecking if numbers are divisible by 3 or 5:')
fizz_buzz_counter(1, 100, 3, 5)
```
